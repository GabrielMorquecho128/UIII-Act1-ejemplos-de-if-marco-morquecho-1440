#### ejemplo de if

```python
# 1. Ejemplo de if simple con operadores relacionales
# Este tipo de estructura ejecuta un bloque de código solo si la condición se cumple

edad = 20  # Asigna el valor 20 a la variable edad [4]
if edad >= 18:  # Verifica si edad es mayor o igual a 18 (operador relacional) [5]
    print("Eres mayor de edad.")  # Muestra el mensaje si la condición es True [6]
```

```python
# 2. Ejemplo de if-else con operadores aritméticos y relacionales
# Se utiliza para elegir entre dos opciones exclusivas
 # En este caso, determinaremos si un número es par utilizando el operador de módulo %

numero = 7  # Define una variable con un número entero [7]
if numero % 2 == 0:  # Calcula el resto de dividir por 2 y compara si es igual a 0 [5, 9]
    print("El número es par.")  # Se ejecuta si el residuo es 0
else:  # Si la condición anterior fue falsa [1]
    print("El número es impar.")  # Se ejecuta si el residuo no es 0 [10]
```

```python
# 3. Ejemplo de if-elif-else con operadores lógicos y relacionales
# Esta estructura es ideal para manejar múltiples condiciones alternativas
# Utilizaremos el operador lógico and para verificar rangos

puntaje = 90  # Valor de prueba para el puntaje
if puntaje >= 90:  # Verifica si es excelente
    print("Calificación: A")
elif puntaje >= 80 and puntaje < 90:  # Verifica si está entre 80 y 89 usando 'and' [12, 13]
    print("Calificación: B")
elif puntaje >= 70 and puntaje < 80:  # Verifica el siguiente rango de notas [13]
    print("Calificación: C")
else:  # Para cualquier puntaje menor a 70 [1]
    print("Calificación: F")
```

```python
# 4. Ejemplo combinado: Año bisiesto
# Este ejemplo utiliza operadores aritméticos (%), relacionales (==, !=) y lógicos (and, or) en una sola expresión compleja

anio = 2024  # Define el año a verificar
# Un año es bisiesto si es divisible por 4 y no por 100, O si es divisible por 400 [6]
if (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0):
    print(f"{anio} es un año bisiesto.")  # Mensaje si es bisiesto [6]
else:
    print(f"{anio} no es un año bisiesto.")  # Mensaje si no lo es [6]
```
